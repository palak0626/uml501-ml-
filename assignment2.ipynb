```python
import pandas as pd

df = pd.read_csv('/content/AWCustomers.csv (2).zip')


df.columns = df.columns.str.strip()

print(df.info())


print(df.head())


for col in df.columns:
    print(col, ":", df[col].nunique())
```

```python
selected_columns = [
    'Gender',
    'MaritalStatus',
    'TotalChildren',
    'NumberChildrenAtHome',
    'Education',
    'Occupation',
    'HomeOwnerFlag',
    'YearlyIncome'
]

df_selected = df[selected_columns].copy()
df_selected.head()
```

```python
from sklearn.preprocessing import MinMaxScaler


df_selected['Gender'] = df_selected['Gender'].map({'M':0, 'F':1})


df_selected['HomeOwnerFlag'] = df_selected['HomeOwnerFlag'].astype(int)


education_mapping = {
    'Partial High School': 1,
    'High School': 2,
    'Partial College': 3,
    'Bachelors': 4,
    'Masters': 5,
    'Doctorate': 6
}
df_selected['Education'] = df_selected['Education'].map(education_mapping)


marital_mapping = {
    'Single': 0,
    'Married': 1,
    'Divorced': 2
}
df_selected['MaritalStatus'] = df_selected['MaritalStatus'].map(marital_mapping)


df_selected = pd.get_dummies(df_selected, columns=['Occupation'], prefix='Occ')


scaler = MinMaxScaler()
df_selected['YearlyIncome'] = scaler.fit_transform(df_selected[['YearlyIncome']])

df_selected.to_csv('AdventureWorks_Preprocessed.csv', index=False)
df_selected.head()
```

# Part III:
Calculating Proximity /Correlation Analysis of two features
Make sure each attribute is transformed in a same scale for numeric attributes and Binarization for each
nominal attribute, and each discretized numeric attribute to standardization. Make sure to apply a correct
similarity measure for nominal (one hot encoding)/binary attributes and numeric attributes respectively.
(a) Calculate Similarity in Simple Matching, Jaccard Similarity, and Cosine Similarity between two
following objects of your transformed input data.
(b) Calculate Correlation between two features Commute Distance and Yearly Income

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import pairwise_distances
import numpy as np


df = pd.read_csv('AdventureWorks_FullyPreprocessed.csv')
df.head()
```

```python

obj1 = df.iloc[0]
obj2 = df.iloc[1]
```

```python

numeric_cols = ['YearlyIncome_std', 'TotalChildren_std', 'NumberChildrenAtHome_std']

binary_cols = [col for col in df.columns if df[col].nunique() == 2 or col.startswith('Occ_') or col.startswith('Marital_')]


num_array = df[numeric_cols].values
bin_array = df[binary_cols].values


num_obj1 = num_array[0].reshape(1, -1)
num_obj2 = num_array[1].reshape(1, -1)

bin_obj1 = bin_array[0].reshape(1, -1)
bin_obj2 = bin_array[1].reshape(1, -1)
```

```python

sm = np.sum(bin_obj1 == bin_obj2) / bin_obj1.shape[1]

intersection = np.sum(np.logical_and(bin_obj1, bin_obj2))
union = np.sum(np.logical_or(bin_obj1, bin_obj2))
jaccard = intersection / union


cosine = np.dot(bin_obj1, bin_obj2.T)[0][0] / (np.linalg.norm(bin_obj1) * np.linalg.norm(bin_obj2))

print("Binary Features Similarity:")
print("Simple Matching:", sm)
print("Jaccard:", jaccard)
print("Cosine:", cosine)
```

```python

cosine_num = np.dot(num_obj1, num_obj2.T)[0][0] / (np.linalg.norm(num_obj1) * np.linalg.norm(num_obj2))
print("\nNumeric Features Cosine Similarity:", cosine_num)
```
